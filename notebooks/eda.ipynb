```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("../data/raw/SampleSuperstore.csv")

# -------------------------------
# 1. Region vs Sales
# -------------------------------
region_sales = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
print("\nRegion vs Sales:\n", region_sales)

region_sales.plot(kind="bar")
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.show()


# -------------------------------
# 2. Category vs Profit
# -------------------------------
category_profit = df.groupby("Category")["Profit"].sum().sort_values(ascending=False)
print("\nCategory vs Profit:\n", category_profit)

category_profit.plot(kind="bar")
plt.title("Total Profit by Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()


# -------------------------------
# 3. Discount vs Profit
# -------------------------------
discount_profit = df.groupby("Discount")["Profit"].mean().sort_index()
print("\nDiscount vs Average Profit:\n", discount_profit)

discount_profit.plot(kind="line", marker='o')
plt.title("Discount vs Average Profit")
plt.xlabel("Discount")
plt.ylabel("Average Profit")
plt.show()
```
